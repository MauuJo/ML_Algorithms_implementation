```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.random.randint(low=0,high=24,size=(20))
x1_sorted = np.sort(x1)

onesArray = [1 for i in range(len(x1))]
hours = np.array([onesArray,x1_sorted])
hours = hours.T

print(hours)
```

```python
ranks = np.random.randint(low=1,high=20,size=(20))

ranks_sorted = np.sort(ranks)[::-1]
ranks = np.array([ranks_sorted])
ranks = ranks.T

print(ranks)
```

```python
#w = np.random.randint(low=0,high=1,size=(2))
w = np.array([[1,1]])
w = w.T

print(w)
```

```python
loss = []
total_epoch = 19000
learning_rate = 0.001
loss_  = 5
epoch = 0
```

```python

while epoch < total_epoch and loss_ > 1:
    p_ranks = np.dot(hours, w)
    errors = ranks - p_ranks
    loss_ = np.mean(errors ** 2)
    loss = np.append(loss,loss_)

    partial = np.dot(hours.T, errors) / len(x1)
    w = w + learning_rate * partial
    epoch +=1

print("Final Loss:", loss[-1])
print("Final Weights:", w)

x_decision = np.linspace(min(hours[:, 1]) - 1, max(hours[:, 1]) + 10, 10)
y_decision = (-w[0][0] - w[1][0] * x_decision) / w[1][0]

plt.scatter(hours[:, 1], ranks)
plt.plot(x_decision, y_decision, label='Decision Boundary', color='red')
plt.show()



plt.plot( loss)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
    
    
        
    
    


                                                                                         




```
