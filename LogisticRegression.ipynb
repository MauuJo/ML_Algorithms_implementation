```python
# Define feature normalization function
def normalize_features(X):
    X_normalized = np.zeros_like(X)
    num_features = X.shape[1]
    
    for i in range(num_features):
        mean = np.mean(X[:, i])
        std = np.std(X[:, i])
        X_normalized[:, i] = (X[:, i] - mean) / std
        
    return X_normalized
```

```python
#sigmoid function

def sigmoidfunc(z):
    result = np.array([])
    for i in range(len(z)):
        #print(z[i][0])
        #print(1/( 1 + np.exp(-z[i][0])))
        result = np.append(result,(1/( 1 + np.exp(-z[i][0]))))

    
    return result
```

```python
#cross entropy calculation function



def error(y, y_p):
    c_error = 0
    epsilon = 1e-15  

    for i in range(len(y)):
        y_p_clip = np.clip(y_p[i][0], epsilon, 1 - epsilon)
        err = y[i][0] * np.log2(1 / y_p_clip) + (1 - y[i][0]) * np.log2(1 - y_p_clip)
        c_error -= err

    return c_error



        
```

```python
#logistic regression function

def logistic_reg(total_epoch, learning_rate,x_, w):

    epoch = 0
    loss = np.array([])
    while epoch < total_epoch and loss_ > 1:
        predicted = sigmoidfunc(y*(np.dot(x_, w)))
        predicted = np.array([predicted])
        predicted = predicted.T
    
        c_error = error(y, predicted)
       
        loss = np.append(loss,c_error)
        s = sigmoidfunc(-y*(np.dot(x_, w)))
        epsilon = 1e-15  
        temp = x_*y
        partial = np.array([0 , 0 , 0])
        
        for i in range(len(temp)):
            s[i] = np.clip(s[i], epsilon, 1 - epsilon)
            partial = partial + temp[i][0]/s[i]
            
        w = w - learning_rate * partial
        epoch +=1

    print("Final Loss:", loss[-1])
    print("Final Weights:", w)

    return w, loss


    
        
```
