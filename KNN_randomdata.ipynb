```python
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
from matplotlib import pyplot as plt
```

```python
x = np.random.randint(0,200, size = (100,2))
x = np.array(x)
print(x)
plt.scatter(x[:,0],x[:,1])

y = np.random.choice(np.arange(len(x)), size=30, replace=False)
y = np.array([y])
y = y.T
a = y[0:10,0]
b = y[10:20,0]
test = y[20:,0]
print(y)
print(a)
print(b)
print(test)
```

```python
A_ = x[a]
aArray = [ 1 for i in range(len(a))]
A = np.array([A_[:,0],A_[:,1],aArray])
A = A.T
print(A)

B_ = x[b]
bArray = [-1 for i in range(len(b))]
B = np.array([B_[:,0],B_[:,1],bArray])
B = B.T
print(B)

TEST = x[test]

index_del = np.concatenate((a,b))
index_del = np.concatenate((index_del,test))
train = np.delete(x,index_del, axis = 0)
print(train)
training = np.concatenate((A,B))
```

```python
plt.scatter(A[:,0],A[:,1], color = 'red')
plt.scatter(B[:,0],B[:,1], color = 'blue')
plt.scatter(train[:,0],train[:,1], color = 'black')
plt.scatter(TEST[:,0],TEST[:,1], color = 'yellow')
plt.show()
```

```python
y_predicted = []

for j in range(len(train)):
    
    distance = np.array([])

    for i in range(len(training)):
        dis = np.sqrt((train[j][0] - training[i][0])**2 + (train[j][1] - training[i][1])**2)
        distance = np.append(distance,dis)
    
    distance_arr = np.array([distance,training[:,-1]])
    distance_arr = distance_arr.T
    
    sorted_indices = np.argsort(distance_arr[:, 0])
    distance = distance_arr[sorted_indices]

    if(distance[0][1] == 1):
        y_predicted = np.append(y_predicted,1)
    else:
        y_predicted = np.append(y_predicted,-1)

    
train = np.array([train[:,0],train[:,1],y_predicted]) 
train = train.T
print(train)
   
plt.scatter(A[:,0],A[:,1], color = 'red')
plt.scatter(B[:,0],B[:,1], color = 'blue')
plt.scatter(train[:,0][train[:,2]==-1],train[:,1][train[:,2]==-1], color = 'blue')
plt.scatter(train[:,0][train[:,2]==1],train[:,1][train[:,2]==1], color = 'red')
plt.scatter(TEST[:,0],TEST[:,1], color = 'yellow')
plt.show()
```

```python

coords = np.concatenate((train,A,B))
print(len(coords))
coords_ = np.delete(coords,2,axis = 1)
vor2 = Voronoi(coords_)

plt.scatter(A[:,0],A[:,1], color = 'red')
plt.scatter(B[:,0],B[:,1], color = 'blue')
plt.scatter(train[:,0][train[:,2]==-1],train[:,1][train[:,2]==-1], color = 'blue')
plt.scatter(train[:,0][train[:,2]==1],train[:,1][train[:,2]==1], color = 'red')
plt.scatter(TEST[:,0],TEST[:,1], color = 'yellow')

voronoi_plot_2d(vor2, show_vertices=False, line_colors='k', line_width=1, line_alpha=0.5)

# Assign colors to Voronoi regions based on the colors of the corresponding points
for region, color in zip(vor2.regions, coords[:, 2]):
    if not -1 in region and color != 0:
        polygon = [vor2.vertices[i] for i in region]
        plt.fill(*zip(*polygon), color=plt.cm.jet(color), alpha=0.5)

plt.show()

#fig2 = voronoi_plot_2d(vor2)
#plt.show()
```
