```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
```

```python
random.seed(98103)
n = 30

x = [random.random() for i in range(n)]
x = sorted(x)
x = np.array(x,dtype=np.float32)
print(x)
```

```python
y = np.sin(4*x)
print(y)
```

```python
random.seed(1)
e = [random.gauss(0,1.0/3.0) for i in range(n)]
y = y+e
print(y)
```

```python
y = np.array(y,dtype=np.float32)
print(y)
```

```python
#df function

def poly_features(x,degree):
    df = pd.DataFrame()

    for i in range(degree+1):
        df['x'+str(i)] = x**i

    return df


    
```

```python
df = poly_features(x,4)
print(df)
```

```python
x = np.array(df)
print(x)
```

```python
#loss function

def loss_function(w,x):
    loss = []
    total_epoch = 18000
    learning_rate = 0.001
    loss_  = 5
    epoch = 0

    while epoch < total_epoch and loss_ > 0.1:
        p_y = np.dot(x, w)
        errors = y - p_y
        loss_ = np.mean(errors ** 2)
        loss = np.append(loss,loss_)
    
        partial = np.dot(x.T, errors) / len(x)
        w = w + learning_rate * partial
        epoch +=1

    
    
    return loss,w
    
        
```

```python
#plotting the data

def plot_func(w, loss, x):
    plt.scatter(x[:,1],y)
    x_decision =  [random.random() for i in range(1000)]
    x_decision = np.array(x_decision,dtype=np.float32)

    f_rand = poly_features(x_decision,4)
    f_rand = np.array(f_rand)

    y_decision = np.dot(f_rand,w)

    plt.scatter(x_decision, y_decision, color='red',s=3)
    plt.autoscale()
    
    plt.plot(loss)
    
    plt.show()
    
```

```python
#regression function

def regression_func(x):
    m,n = np.shape(x)
    w = [random.random() for i in range(n)]
    w = np.array(w)

    loss,f_w = loss_function(w,x)
    print("Final Loss:", loss[-1])
    print("Final Weights:", w)

    
    #plot_func(w, loss, x)
    return loss, f_w
    
    
```
